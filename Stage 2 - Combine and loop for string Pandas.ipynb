```python
def opinionnotification(df):
    for index, row in df.iterrows():
        if 'i wanna know your name' in row['text'].lower():
            df.at[index, 'Opinion'] = 'Wants to know name'
        elif 'i want to know your name' in row['text'].lower():
            df.at[index, 'Opinion'] = 'Wants to know name'
        elif 'i wanna know her name' in row['text'].lower():
            df.at[index, 'Opinion'] = 'Wants to know name'
        elif 'i want to know her name' in row['text'].lower():
            df.at[index, 'Opinion'] = 'Wants to know name'
        elif 'just want to know your name' in row['text'].lower():
            df.at[index, 'Opinion'] = 'Wants to know name'
        elif 'just wanna know your name' in row['text'].lower():
            df.at[index, 'Opinion'] = 'Wants to know name'
        elif 'wanted to know your name' in row['text'].lower():
            df.at[index, 'Opinion'] = 'Wants to know name'
        elif 'wanted to know her name' in row['text'].lower():
            df.at[index, 'Opinion'] = 'Wants to know name'
        elif "don't want to know your name" in row['text'].lower():
            df.at[index, 'Opinion'] = "Doesn't want to know name"
        elif "don't wanna know your name" in row['text'].lower():
            df.at[index, 'Opinion'] = "Doesn't want to know name"
        elif "don't know your name" in row['text'].lower():
            df.at[index, 'Opinion'] = "Reasonably concerned"    
        elif "did not know your name" in row['text'].lower():
            df.at[index, 'Opinion'] = "Reasonably concerned"
        else:
            df.at[index, 'Opinion'] = "Unknown"
    return df
```
